```python
month_end_index = intra_turnover[intra_turnover['index'] == month_end[1]].index
intra_return.iloc[month_end_index[0]-19:month_end_index[0]]
```

```python
intra_mom = pd.DataFrame(index=month_end, columns=stocks)
ov_mom = pd.DataFrame(index=month_end, columns=stocks)
new_mom = pd.DataFrame(index=month_end, columns=stocks)
for date in month_end:
    month_end_index = intra_turnover[intra_turnover['index'] == date].index
    intra_ret = intra_return.iloc[month_end_index[0]-19: month_end_index[0]+1].set_index('index')
    ov_ret = ov_return.iloc[month_end_index[0]-19: month_end_index[0]+1].set_index('index')
    intra_turn = intra_turnover.iloc[month_end_index[0]-19: month_end_index[0]+1].set_index('index')
    yesterday_turn = turnover_rate.loc[month_end_index[0]-20: month_end_index[0]].set_index('index')
    intra_dist = ((intra_turn.shape[0] - intra_turn.isna().sum()) / 5).apply(int)
    intra_rank = intra_turn.rank(method='first')
    intra_part1 = intra_ret[intra_rank <= intra_dist].mean()
    intra_part1_std = (intra_part1 - intra_part1.mean()) / intra_part1.std()
    intra_part2 = intra_ret[intra_rank > intra_turn.shape[0]-intra_dist].mean()
    intra_part2_std = (intra_part2 - intra_part2.mean()) / intra_part2.std()
    intra_mom.loc[date] = -intra_part1_std + intra_part2_std
    intra_mom.loc[date] = (intra_mom.loc[date] - intra_mom.loc[date].mean()) / intra_mom.loc[date].std()
    ov_dist = ((yesterday_turn.shape[0] - yesterday_turn.isna().sum()) / 5).apply(int)
    ov_rank = yesterday_turn.rank(method='first')
    ov_part1 = ov_ret[ov_rank <= ov_dist].mean()
    ov_part1_std = (ov_part1 - ov_part1.mean()) / ov_part1.std()
    ov_part2 = ov_ret[ov_rank > yesterday_turn.shape[0]-ov_dist].mean()
    ov_part2_std = (ov_part2 - ov_part2.mean()) / ov_part2.std()
    ov_mom.loc[date] = ov_part1_std - ov_part2_std
    ov_mom.loc[date] = (ov_mom.loc[date] - ov_mom.loc[date].mean()) / ov_mom.loc[date].std()
    new_mom.loc[date] = intra_mom.loc[date] + ov_mom.loc[date]
```

```python
from sklearn import linear_model

res_mom = pd.DataFrame(index=month_end, columns=stocks)
for date in month_end:
    model = linear_model.LinearRegression()
    model.fit(old_mom.loc[date].fillna(0).values.reshape(-1, 1), new_mom.loc[date].fillna(0).values.reshape(-1, 1))
    res_mom.loc[date] = new_mom.loc[date].values - old_mom.loc[date].values * model.coef_
```
